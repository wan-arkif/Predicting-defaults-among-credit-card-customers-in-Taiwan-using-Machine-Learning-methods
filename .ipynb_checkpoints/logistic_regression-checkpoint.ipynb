* แก้ไข Index ให้เป็น ID เพื่อให้สะดวกต่อการวิเคราะห์

```python
df.set_index("ID", inplace = True)
```

* ตรวจสอบค่า Null ของข้อมูล

```python
df.info()
```

จากการตรวจสอบข้างต้นไม่พบข้อมูล Null ในตัวแปรใดๆของข้อมูลการผิดนัดชำระของลูกค้าบัตรเครดิตในไต้หวัน

* ตรวจสอบข้อมูลในตัวแปร EDUCATION และ MARRIAGE

```python
df['EDUCATION'].value_counts().sort_index(ascending = True)
```

```python
df['MARRIAGE'].value_counts().sort_index(ascending = True)
```

จากตัวแปรอิระ EDUCATION พบค่าที่ไม่สามารถอธิบายความหมายของข้อมูล ได้แก่ 0, 5, 6 และตัวแปรอิระ MARRIAGE พบค่าที่ไม่สามารถอธิบายความหมายของข้อมูล ได้แก่ 0

* ซึ่งผู้วิจัยได้ทำการแก้ไขปัญหาดังกล่าว โดยแทนค่า 0, 5, 6 ที่มาจากตัวแปรอิระ EDUCATION เป็น 'others' และแทนค่า 0 ที่มาจากตัวแปรอิระ MARRIAGE เป็น 'others'

```python
def impute_education(cols):
    EDUCATION = cols.iloc[0] 

    if EDUCATION not in [1, 2, 3, 4]:
        return 4
    else:
        return EDUCATION
```

```python
df['EDUCATION'] = df[['EDUCATION']].apply(impute_education,axis=1)
```

```python
def impute_marriage(cols):
    MARRIAGE = cols.iloc[0] 

    if MARRIAGE not in [1, 2, 3]:
        return 3
    else:
        return MARRIAGE
```

```python
df['MARRIAGE'] = df[['MARRIAGE']].apply(impute_marriage,axis=1)
```

* เปลี่ยนชื่อตัวแปรตาม default payment next month เป็น DEFAULT เพื่อให้สะดวกในการวิเคราะห์ข้อมูล

```python
df = df.rename(columns = {'default payment next month': 'DEFAULT'})
```

* เปลี่ยนชื่อตัวแปรอิสระ PAY_0 เป็น PAY_1 เพื่อให้สอดคล้องกับข้อมูลดังกล่าว

```python
df = df.rename(columns = {'PAY_0': 'PAY_1'})
```

```python
x = df.drop('DEFAULT',axis=1)
```

```python
y = df['DEFAULT']
```

```python
smote_enn = SMOTEENN(random_state=101)
x_resampled, y_resampled = smote_enn.fit_resample(x, y)

print(Counter(y_resampled))
```

* แบ่งข้อมูลที่ใช้ในการวิเคราะห์ออกเป็น 2 ส่วน โดยในส่วนแรกเป็นข้อมูลที่ใช้ในการสร้างโมเดล(Traning Data) ซึ่งเป็นร้อยละ 70 ของข้อมูลทั้งหมด และในส่วนที่สองเป็นข้อมูลที่ใช้ในการทดสอบประสิทธิภาพในการทำนายค่าของโมเดล(Testing Data) ซึ่งเป็นร้อยละ 30 ของข้อมูลทั้งหมด

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.30, random_state=101, stratify = y_resampled)
```

* เนื่องจากข้อมูลในแต่ละฟีเจอร์มีขนาดที่แตกต่างกันมาก ผู้วิจัยจึงทำการปรับขนาดของฟีเจอร์โดยใช้วิธี StandardScaler

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns, index=x_train.index)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns, index=x_test.index)
```

# 2. สร้างโมเดล Logistic Regression 

* ความหมายของไฮเปอร์พารามิเตอร์แต่ละตัวในโมเดล Rogistic Regression

1.) C (Inverse of Regularization Strength) คือ ค่าที่ใช้ในการควบคุมความแรงของ Regularization(ค่าเล็ก = Regularization มาก)

2.) penalty คือ ประเภทของ Regularization ได้แก่ l1, l2, elasticnet และ none

3.) solver คือ อัลกอริธึมที่ใช้ในการหาค่าพารามิเตอร์ ได้แก่ lbfgs, liblinear, newton-cg, sag และ saga

4.) max_iter คือ จำนวนรอบสูงสุดในการไล่ค่าพารามิเตอร์

5.) tol คือ ค่าความคลาดเคลื่อนขั้นต่ำที่ยอมรับได้

6.) class_weight คือ ถ่วงน้ำหนักคลาส (balanced หรือ กำหนดเองเป็น dict)

```python
log_model = LogisticRegression()
```

```python
log_model.fit(x_train_scaled, y_train)
```

```python
predict = log_model.predict(x_test_scaled)
```

```python
print(classification_report(y_test,predict))
```

# 3. ปรับปรุงแบบจำลอง Rogistic Regression โดยใช้ Grid Search ในการค้าหาไฮเปอร์พารามิเตอร์ที่เหมาะสมมากที่สุด

```python
param_grid = [
    {'penalty': ['l1', 'l2'], 'solver': ['liblinear'], 'C': [0.01, 0.1, 1, 10, 100], 'max_iter': [500]}, 
    {'penalty': ['elasticnet'], 'solver': ['saga'], 'C': [0.01, 0.1, 1, 10, 100], 'l1_ratio': [0.1, 0.5, 0.9], 'max_iter': [500]}, 
    {'penalty': ['l2'], 'solver': ['saga'], 'C': [0.01, 0.1, 1, 10, 100], 'max_iter': [500]}
]
```

```python
grid_search = GridSearchCV(log_model, param_grid, cv=5, scoring='accuracy', error_score='raise', refit=True, verbose=2)
grid_search.fit(x_train_scaled, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

# 4. ทดสอบประสิทธิภาพในการทำนายค่าของโมเดล Logistic Regression

* ทดสอบประสิทธิภาพในการทำนายโอกาสการผิดนัดชำระหนี้ของลูกค้าบัตรเครดิตในไต้หวันโดยโมเดล Logistic Regression จากชุดข้อมูล Testing Data

```python
predict = grid_search.predict(x_test_scaled)
```

```python
print(classification_report(y_test,predict))
```

จากผลการทดสอบประสิทธิภาพในการทำนายโอกาสการผิดนัดชำระหนี้ของลูกค้าบัตรเครดิตในไต้หวันโดยโมเดล Logistic Regression พบว่ามีค่า Accuracy เท่ากับ 0.77 หมายความว่าโมเดลมีประสิทธิภาพในการทำนายที่แม่นยำ 77%

แต่ถ้าหากวิเคราะห์จากค่า Precision และค่า Recall ของโอกาสที่ลูกค้าจะผิดนัดชำระหนี้บัตรเครดิต และโอกาสที่ลูกค้าจะไม่ผิดนัดชำระหนี้บัตรเครดิต พบว่าโมเดล Logistic Regression มีความสามารถในการทำนายค่าของกลุ่มลูกค้าที่ไม่ผิดนัดชำระหนี้บัตรเครดิตได้แม่นยำ 73% และมีความสามารถในการ Recall ของกลุ่มลูกค้าที่ไม่ผิดนัดชำระหนี้บัตรเครดิต 63% และมีความสามารถในการทำนายค่าของกลุ่มลูกค้าที่มีการผิดนัดชำระหนี้บัตรเครดิตได้แม่นยำ 79% แต่มีความสามารถในการ Recall ของกลุ่มลูกค้าที่มีการผิดนัดชำระหนี้บัตรเครดิตเพียง 86%
